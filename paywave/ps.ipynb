```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
models = ["Multinomial NB", "Logistic Regression", "KNN", "Ada Boosting", "Gradient Boosting"]
accuracy = [95, 96, 95, 88, 88]

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.barplot(x=accuracy, y=models, palette="viridis")

# Adding percentages on top of the bars
for i in range(len(models)):
    ax.text(accuracy[i], i, f'{accuracy[i]}%', va='center', fontsize=12, color='black', ha='right')

# Adding labels and title
plt.xlabel("Accuracy")
plt.title("Model Accuracy Comparison")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the category labels
categories = ['Benign URLs', 'Defacement URLs', 'Phishing URLs', 'Malware URLs']

# Define the corresponding percentages
percentages = [65.67, 14.81, 14.46, 4.99]

# Define the colors for the pie chart
colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'lightyellow']

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title and display the chart
plt.title("Distribution of Malicious URL Categories")
plt.show()

# Display the total number of URLs
total_urls = 651191
print(f"Total number of URLs: {total_urls}")
```
